```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Send a Request to the Web Page
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
html_data = requests.get(url).text

```

```python
soup = BeautifulSoup(html_data, "html.parser")
soup.find_all('table')
```

```python
gme_revenue = pd.DataFrame(columns = ['Date', 'Revenue'])

for row in soup.find_all("tbody")[1].find_all("tr"):
    col = row.find_all("td")
    date = col[0].text
    revenue = col[1].text.replace("$", "").replace(",", "")
    
    gme_revenue = gme_revenue.append({"Date": date, "Revenue": revenue}, ignore_index = True)
```

```python
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
gme_revenue.tail()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph():
    # Fetching Tesla stock data using yfinance
    ticker = 'TSLA'
    tesla_data = yf.download(ticker, start='2020-01-01', end='2023-01-01')  # Adjust dates as needed
    
    plt.figure(figsize=(10, 6))
    plt.plot(tesla_data.index, tesla_data['Close'], label='Tesla Stock (TSLA)')
    
    plt.title('Tesla Stock Price (Close)')
    plt.xlabel('Date')
    plt.ylabel('Stock Price ($)')


# Calling the function to generate the plot
make_graph()
```
